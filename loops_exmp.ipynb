```python
#1.2 Print the numbers which are multiple of 7 from 1 to 1000.
    #- Expected Output: 7,14,21,28,35,42,49,56,63,70,77,84,91,98...
```

```python
#method1
for number in range(1,1000):
    mult = 7*number
    if mult>1000:
        break
    else:
        print(mult)
    
```

```python
number = 1
```

```python
while(number!=1000):
    mult = number*7
    if mult>1000:
        break
    else:
        print(mult)
    number =number+1
```
